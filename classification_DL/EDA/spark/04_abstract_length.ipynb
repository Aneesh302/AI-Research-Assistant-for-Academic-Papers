```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,split,count, size, explode, length, when

spark = SparkSession.builder \
    .appName("arxiv_eda") \
    .getOrCreate()

df = spark.read.json("../../../data/arxiv-metadata-oai-snapshot.json")
```

Character Length

```python
len_df = df.withColumn("abstract_len", length(col("abstract")))
```

```python
len_df.select("abstract_len").summary().show()
```

```python
len_df.approxQuantile("abstract_len", [0.5, 0.9, 0.95, 0.99], 0.01)
```
