```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,year,to_date,col,split,count, size, explode, length, when

spark = SparkSession.builder \
    .appName("arxiv_eda") \
    .getOrCreate()

df = spark.read.json("../../../data/arxiv-metadata-oai-snapshot.json")
```

```python
df.withColumn("num_versions", size(col("versions"))) \
  .groupBy("num_versions").count().orderBy("num_versions").show()
```
