```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,year,to_date,col,split,count, size, explode, length, when

spark = SparkSession.builder \
    .appName("arxiv_eda") \
    .getOrCreate()

df = spark.read.json("../../../data/arxiv-metadata-oai-snapshot.json")
```

```python
spark.catalog.clearCache()
spark.stop()
```

```python
df_time = df.withColumn(
    "year",
    year(to_date(col("update_date")))
)
```

```python
year_stats = df_time.groupBy("year").count().orderBy("year")
year_stats.show()
```

```python
cat_year = df_time \
    .withColumn("category", explode(split(col("categories"), " "))) \
    .groupBy("year", "category").count()

cat_year.orderBy(col("year").desc(), col("count").desc()).show()

```
