```python
import json
import pandas as pd

def stream_arxiv_json(file_path, n=10000):
    
    with open(file_path, "r") as f:
        for i, line in enumerate(f):
            if i >= n:
                break
            yield json.loads(line)


# Use generator to load first 1000 rows
data_gen = stream_arxiv_json("../data/arxiv-metadata-oai-snapshot.json", n=1000)

# Convert to DataFrame
df = pd.DataFrame(data_gen)

print(df.shape)
```

```python
def batch_stream(file_path, batch_size=1000):
    batch = []
    with open(file_path) as f:
        for line in f:
            batch.append(json.loads(line))
            if len(batch) == batch_size:
                yield pd.DataFrame(batch)
                batch = []

# get first batch only
df = next(batch_stream("arxiv.json", 1000))
```
